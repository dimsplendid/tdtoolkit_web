```python
# make graph transparents
from PIL import Image

img = Image.open("image-test.png")
img = img.convert('RGBA')
x, y = img.size 

for i in range(x):
    for j in range(y):
        color = img.getpixel((i,j))
        color = color[:-1] + (90, )
        img.putpixel((i, j), color)

img
```
